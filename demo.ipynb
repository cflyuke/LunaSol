```python
"""
orbit类的最简单用法
复杂用法在后面

默认的参数：
    解法：'RK45'
    rtol: 1e-6
    atol: 1e-6
    天体：日月地
    动力学方程：三体问题方程
"""
from LunaSol.Orbit import Orbit
orbit = Orbit()
orbit.visualize(startyear=2025, endyear=2050) 
```

```python
""""
演示LunarEclipse最简单的用法，你也可以直接使用lunar.pipeline()自动帮你完成以下事情。
lunarEclipse完全类似，不再重复。
"""

from LunaSol.LunarEclipse import LunarEclipse
lunar = LunarEclipse()
files = lunar.predict_eclipse(startyear=2025, endyear=2050)  #进行预测，返回预测结果文件路径列

import pandas as pd
df = pd.read_csv(files[0])
display(df.head(10)) #展示一下结果
lunar.visualize_error(startyear=2025, endyear=2050, predict_files=files) #误差的可视化
lunar.visualize_orbit(2025, files[0]) #月食路径的可视化
```

```python
"""
接下来展示更加复杂的用法，你可以自己定义各类参数
下面加入了木星进行考虑
你还可以添加更多的星体，只要你加入的kernelfile支持即可
"""
orbit = Orbit(
    bodies=['EARTH', 'MOON', 'SUN', 'JUPITER'],
    masses={'EARTH': 5.972e24,'MOON': 7.342e22, 'SUN': 1.989e30, 'JUPITER': 1.898e27},
    kernel_files=['data/de442.bsp', 'data/earth_200101_990827_predict.bpc', 'data/naif0012.tls', 'data/jup346.bsp']
)
count = 1
for time, state in orbit.predict(startyear=2030, endyear=2035, dynamics_func='standard', method='DOP853', rtol=1e-8, atol=1e-6):
    print(time)
    print(state)
    count +=1
    if count > 3:
        break
```

```python
"""
多种参数下的可视化分析
"""
orbit.visualize_error(startyear=2025, endyear=2050,
                      dynamics_funcs=['standard'],
                      methods=['RK45', 'DOP853'],
                      rtols=[1e-6, 1e-8],
                      atols=[1e-6, 1e-8],
                      stepsize=600)
"""
不难看出各种各样配置下的误差差距非常小，而且各个星体的误差呈现出按自身绕太阳旋转的周期变换
"""
```

```python
"""
下面来展示LunarEclipse的用法
这次我们加入金星来考虑
换一种求解器来预测日食轨迹
"""
from LunaSol.LunarEclipse import LunarEclipse
lunar = LunarEclipse(
    reference_file_path = 'data/lunar_eclipse_data.csv',
    bodies=['EARTH', 'MOON', 'VENUS', 'SUN'],
    masses={'EARTH': 5.972e24,'MOON': 7.342e22, 'VENUS': 4.867e24, 'SUN': 1.989e30},
)
files = lunar.predict_eclipse(startyear=2025, endyear=2075,
                              time_step = 1,
                              use_official_data=True,
                              dynamics_funcs=['standard'],
                              methods=['DOP853'],
                              rtols=[1e-6, 1e-8],
                              atols=[1e-6])
lunar.visualize_error(startyear=2025, endyear=2075, predict_files=files)
lunar.visualize_orbit(year=2025, filepath=files[1], degree=1, time_step=180,
                    use_official_data=False, method='DOP853', rtol=1e-6, atol=1e-6)  ## 使用这种微分方程求解器配置来预测轨道
```
