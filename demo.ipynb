```python
"""
orbit类的最简单用法
复杂用法在后面

默认的参数：
    解法：'RK45'
    rtol: 1e-6
    atol: 1e-6
    天体：日月地
    动力学方程：三体问题方程
"""
from LunaSol.Orbit import Orbit
orbit = Orbit()
orbit.visualize_error(startyear=2025, endyear=2050) 
```

```python
""""
演示LunarEclipse最简单的用法，你也可以直接使用lunar.pipeline()自动帮你完成以下事情。
lunarEclipse完全类似，不再重复。
"""

from LunaSol.LunarEclipse import LunarEclipse
lunar = LunarEclipse()
files = lunar.predict_eclipse(startyear=2025, endyear=2050)  #进行预测，返回预测结果文件路径列

import pandas as pd
df = pd.read_csv(files[0])
display(df.head(10)) #展示一下结果
lunar.visualize_error(startyear=2025, endyear=2050, predict_files=files) #误差的可视化
#lunar.visualize_orbit(2025, files[0]) #月食路径的可视化
```

```python
"""
接下来展示更加复杂的用法，你可以自己定义各类参数

下面的Orbit:
    1. 考虑的天体（加入木星和金星，你也可以加入其他星体）：'EARTH', 'MOON', 'SUN', 'JUPITER', 'VENUS'
    2. 各种参数配置：
        - dynamics_funcs=['standard'],
        - methods=['RK45', 'DOP853'],
        - rtols=[1e-6, 1e-8],
        - atols=[1e-6, 1e-8]

"""
from LunaSol.Orbit import Orbit
orbit = Orbit(
    bodies=['EARTH', 'MOON', 'SUN', 'JUPITER', 'VENUS'],
    masses={'EARTH': 5.972168e24,'MOON': 7.346303e22, 'SUN': 1.988470e30, 'JUPITER': 1.898187e27, 'VENUS': 4.867468e+24},
    kernel_files=['data/de442.bsp', 'data/earth_200101_990827_predict.bpc', 'data/naif0012.tls', 'data/jup346.bsp']
)

"""
多种参数下的可视化分析
"""
orbit.visualize_error(startyear=2025, endyear=2050,
                      dynamics_funcs=['standard'],
                      methods=['RK45', 'DOP853'],
                      rtols=[1e-6, 1e-8],
                      atols=[1e-6, 1e-8],
                      stepsize=600)
```

```python
"""
Solar eclipe
接下来我们进行日食的预测
我们先来考虑，
    1. 只有日月地三者
    2. 各种微分方程配置的情况
        - dynamics_funcs=['standard'],
        - methods=['RK45', 'DOP853'],
        - rtols=[1e-6, 1e-8],
        - atols=[1e-6, 1e-8] 
"""
from LunaSol.SolarEclipse import SolarEclipse
solar = SolarEclipse(
    reference_file_path = 'data/Lunar_eclipse_data.csv',
    bodies=['EARTH', 'MOON', 'SUN'],
    masses={'EARTH': 5.972168e24,'MOON': 7.346303e22, 'SUN': 1.988470e30}
)
files = solar.predict_eclipse(startyear=2025, endyear=2075,
                              time_step = 1,
                              use_official_data=True,
                              dynamics_funcs=['standard'],
                              methods=['RK45', 'DOP853'],
                              rtols=[1e-6, 1e-8],
                              atols=[1e-6, 1e-8])
solar.visualize_error(startyear=2025, endyear=2075, predict_files=files)
```

```python
"""
Solar eclipse with JUPITER and VENUS
考虑其他天体的情况：['EARTH', 'MOON', 'SUN', 'JUPITER', 'VENUS']
配置：使用上面精度最佳的配置
    - dynamics_funcs=['standard'],
    - methods=['DOP853'],
    - rtols=[1e-8],
    - atols=[1e-8] 
"""
from LunaSol.SolarEclipse import SolarEclipse
solar_5_bodies = SolarEclipse(
    reference_file_path = 'data/solar_eclipse_data.csv',
    bodies = ['EARTH', 'MOON', 'SUN', 'JUPITER', 'VENUS'],
    masses = {'EARTH': 5.972168e24 , 'MOON' : 7.346303e22, 'SUN' : 1.988470e30, 'JUPITER': 1.898187e27, 'VENUS': 4.867468e24},
    output_dir = 'output_with_JUPITER_VENUS'
)
files = solar_5_bodies.predict_eclipse(startyear=2025, endyear=2075,
                              time_step = 1,
                              use_official_data=False,
                              dynamics_funcs=['standard'],
                              methods=['DOP853'],
                              rtols=[1e-8],
                              atols=[1e-8])
solar_5_bodies.visualize_error(startyear=2025, endyear=2075, predict_files=files)
```

```python
"""
Lunar eclipe
接下来我们进行月食的预测
我们先来考虑，
    1. 只有日月地三者
    2. 各种微分方程配置的情况
        - dynamics_funcs=['standard'],
        - methods=['RK45', 'DOP853'],
        - rtols=[1e-6, 1e-8],
        - atols=[1e-6, 1e-8] 
"""
from LunaSol.LunarEclipse import LunarEclipse
lunar = LunarEclipse(
    reference_file_path = 'data/Lunar_eclipse_data.csv',
    bodies=['EARTH', 'MOON', 'SUN'],
    masses={'EARTH': 5.972168e24,'MOON': 7.346303e22, 'SUN': 1.988470e30}
)
files = lunar.predict_eclipse(startyear=2025, endyear=2075,
                              time_step = 1,
                              use_official_data=True,
                              dynamics_funcs=['standard'],
                              methods=['RK45', 'DOP853'],
                              rtols=[1e-6, 1e-8],
                              atols=[1e-6, 1e-8])
lunar.visualize_error(startyear=2025, endyear=2075, predict_files=files)
```
